#### 1. Combine the data of 10 companies into one csv file by adding a new dimension “Company”.
Refer to Phase 2 of Task 1 where this was already completed, the data was merged and added a column "Stock Name". Here we will simply change the column "Stock Name" to "Company".

```python
import pandas as pd
import numpy as np

all_data = pd.read_csv('data/merged_data.csv')
all_data = all_data.rename(columns={'Stock Name': 'Company'})
# Drop unneeded index column
all_data = all_data.drop('Unnamed: 0', axis = 1)
all_data
```

#### 2. Only keep the dimensions: Date, Open, Close, Volume, Company; drop the other dimensions.

We need to drop "High", "Low" and "Adj Close"

```python
all_data = all_data.drop(['High', 'Low', 'Adj Close'], axis=1)
all_data
```

#### 3. Sort the data on date. Now you will have 10 rows of data per day where each row represents a different company.

```python
all_data = all_data.sort_values(by=['Date', 'Company'])
all_data
```

#### 4. Add 10 new dimensions to the data, 1 for each company. You can name these dimensions same as the company name.

```python
transformed_data = pd.DataFrame(0, index = range(252), columns=[ 'Date',
                                                        'AAPL', 
                                                         'AMD', 
                                                        'AMZN',
                                                        'GOOG',
                                                         'IBM',
                                                        'INTC',
                                                        'LCID',
                                                        'META',
                                                        'MSFT',
                                                        'TSLA'])
transformed_data
```

#### 5. Merge the 10 rows of each day in the following manner:
a. If the price of a company has gone up i.e., Close-Open is positive then copy the volume value in the dimension of that company, the one that you added in step 4. Otherwise add 0. <br>
b. In the end you will have only 1 row per date and the 10 dimensions added in step 4 will have either 0 or volume-value indicating whether the price of this stock has gone down or up (in this case how many units sold). <br>
c. Drop the Open, Close, Volume and Company dimensions.

```python
# Creating copy of dataframe to store transformed data
all_data_copy = all_data.copy()

for index, row in all_data_copy.iterrows():
    if row['Close'] - row['Open'] <= 0:
        all_data_copy.at[index, 'Volume'] = 0

all_data_copy = all_data_copy.drop(columns = ['Open', 'Close'], axis = 1)
all_data_copy = all_data_copy.reset_index(drop=True)
all_data_copy
```

```python
import math

# Fully transform data
for index, row in all_data_copy.iterrows():
    row_num = math.floor(index/10)
    col_num = (index % 10) + 1
    transformed_data.iloc[row_num, col_num] = row['Volume']

for x in range(252):
    transformed_data.iloc[x, 0] = all_data_copy.iloc[x*10, 0]

transformed_data
```

This generated the transformed data that we are going to use for further analysis in the next tasks.

## Task 1: Association Analysis
Our objective is to identify which stocks/companies go up together. Following this objective, answer the questions below.

#### Find the 10 most frequent item-sets. The length/size of these item-sets must be at least 2. Compute the support for these item-sets.

The following libraries will be the tools used for finding frequent item-sets and association rules.

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

Now the data needs to be transformed to use the Apriori algorithm. The following steps will encode the data so that the mlxtend library can use the apriori algorithm. The encoding simply will transform the data where any positive values will become value "True/1". If zero, it will remain "False/0". We will also drop date as we are simply finding which stocks move up in price regardless of date. Once all positive values are transformed to ones, now the apriori algorithm can take the data as an input can detect which stocks move up in price with each other.

```python
tr = transformed_data.copy()
tr = tr.drop(['Date'], axis=1)
for row_index, row in tr.iterrows():
    for col_index, value in row.items():
        if value > 0:
            tr.loc[row_index, col_index] = 1
tr
```

Now that the data has been transformed, Apriori can be applied to the data.

```python
frequent_itemsets = apriori(tr, min_support=0.4, use_colnames=True)
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)
frequent_itemsets = frequent_itemsets.iloc[:20]
frequent_itemsets
```

Since the size of these items sets needs to be at size 2, the first 10 rows of the output above will be omitted and the next 10 will be considered the 10 most frequent item-sets.

```python
frequent_itemsets_pruned = frequent_itemsets.iloc[10:20]
frequent_itemsets_pruned
```

The support is listed beside the itemsets. As observed they are all around a support of 0.44 to 0.41.

So from the frequent itemset list, we can state that these stocks move up in price with each other.

## Task 2: Clustering Analysis

For this task, data from Phase 2 Task 6 will be used.

```python
all_data_task2 = pd.read_csv('../Phase 2/Data/merged_data.csv')
all_data_task2 = all_data_task2.rename(columns={'Stock Name': 'Company'})
all_data_task2 = all_data_task2.drop('Unnamed: 0', axis = 1)
all_data_task2
```

#### Perform any necessary preprocessing steps. Add explanation to your report.

First preprocessing is neeeded, by binarizing "Date" and "Company".

```python
all_data_task2_copy = all_data_task2.copy()
# Preprocessing for Date
for index, row in all_data_task2_copy.iterrows():
    new_str = row['Date'].replace('-', '')
    all_data_task2_copy['Date'] = all_data_task2_copy['Date'].replace(row['Date'], int(new_str))
all_data_task2_copy
```

```python
#Preprocessing for Company
company_dict = {
    'AAPL': 0,
    'AMD': 1,
    'AMZN': 2,
    'GOOG': 3,
    'IBM': 4,
    'INTC': 5,
    'LCID': 6,
    'META': 7,
    'MSFT': 8,
    'TSLA': 9,
}
for index, row in all_data_task2_copy.iterrows():
    str_int= company_dict[row['Company']]
    all_data_task2_copy['Company'] = all_data_task2_copy['Company'].replace(row['Company'], str_int)
all_data_task2_copy
```

#### Perform the k-means clustering of your data with k=5 and 7. Which value of k produces the least SSE (sum of squared errors)?

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Performing k-means clustering with k=5.

```python
kmeans5 = KMeans(n_clusters=5).fit(all_data_task2_copy)
centroids5 = kmeans5.cluster_centers_
print(centroids5)
```

Performing k-means clustering with k=7.

```python
kmeans7 = KMeans(n_clusters=7).fit(all_data_task2_copy)
centroids7 = kmeans7.cluster_centers_
print(centroids7)
```

Next step is to find the Sum of Squared Errors for k=5 and k=7. For this we can use the inertia_ function to obtain the SSE.

For k=5 ...

```python
kmeans5.inertia_
```

For k=7 ...

```python
kmeans7.inertia_
```

As observed, k=7 has a lower sum of squares error, meaning it better models the regression. <br>
